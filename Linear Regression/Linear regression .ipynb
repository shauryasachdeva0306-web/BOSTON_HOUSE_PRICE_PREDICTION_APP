```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbs
import plotly.express as px
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
year=[2017,2018,2019,2020,2021,2022,2023,2024,2025,2026]
price=[520000,540000,560000,580000,600000,630000,640000,660000,690000,700000]
```

```python
train_x,test_x,train_y,test_y= train_test_split(year,price,test_size=0.2)
```

```python
train_x=np.array(train_x).reshape(8,1)
```

```python
train_y=np.array(train_y).reshape(8,1)
```

```python
test_x=np.array(test_x).reshape(2,1)
```

```python
test_y=np.array(test_y).reshape(2,1)
```

```python
lr=LinearRegression()
```

```python
lr.fit(train_x,train_y)
```

```python
m=lr.coef_
```

```python
c= lr.intercept_
```

```python
h= test_x [:1]
```

```python
(m*h)+c
```

# to check the performance of the model

```python
pred=lr.predict(test_x)
```

```python
MSE= mean_squared_error(test_y,pred)
```

```python
RMSE= MSE**0.5
```

```python
R_square= r2_score(test_y,pred)
```

```python
MAE= mean_absolute_error(test_y,pred)
```

```python
print(MSE) 
print(RMSE)
print(R_square) # how much we are improving by using linear model instead of doing prediction on average
print(MAE)
```

```python
1-((1-R_square)*(len(year)-1))/(len(year)-1-1)    # adjusted r_squared value
```

```python
L1= Ridge()
L2= Lasso()
EN= ElasticNet()
```

```python
L1.fit(train_x,train_y)
L2.fit(train_x,train_y)
EN.fit(train_x,train_y)
```

```python
print(L1.score(train_x,train_y))
print(L2.score(train_x,train_y))
print(EN.score(train_x,train_y))
```

```python
print(L1.score(test_x,test_y))
print(L2.score(test_x,test_y))
print(EN.score(test_x,test_y))
```
